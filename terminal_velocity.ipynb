```python
import numpy as np 
import matplotlib.pyplot as pt
```

## Parameters

```python
g = 9.8 
h = 100000
y = [h]
v = 0
vy = [0]
dt = 0.001
k = 0.05
steps = 0
mx_steps = 10000
time = [0]
t = 0
```

## Simulation 

```python
while h >= 0 and steps < mx_steps:
     
    #acceleration 
    a = -g-k*(v) * abs(v)
    
    #velocity update
    v = v + a * dt
    vy.append(abs(v))

    #position update
    h = h + v * dt
    y.append(h)


    t += dt
    steps += 1

    time.append(t)

    if steps == mx_steps:
        print("ValueError: there is bug in code ")
        break






```

```python

pt.plot(time, vy)
pt.xlabel("time")
pt.ylabel("velocity")

pt.title("velocity vs time ")
pt.grid()
pt.show()
```
